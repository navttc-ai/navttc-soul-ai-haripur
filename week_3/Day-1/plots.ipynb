## 1. Univariate Plots

Univariate plots analyze **one variable** at a time.
- Common Plots: Histogram, Boxplot, KDE plot.
- **Purpose**: Understand distribution, spread, and outliers.
- **Attributes**: `bins`, `kde`, `color`, `alpha`, etc.

```python
# Simple Example: Histogram
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title('Simple Histogram')
plt.show()
```

```python
# Detailed Example: Histogram with KDE
import seaborn as sns

data = np.random.normal(loc=50, scale=10, size=500)
sns.histplot(data, bins=20, kde= True, color='green')
plt.title('Detailed Histogram with KDE')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python

# Boxplot
plt.boxplot(data)
plt.title("Box Plot (Matplotlib)")
plt.ylabel("Values")
plt.show()
```

## 2. Bivariate Plots

Bivariate plots analyze **two variables together**.
- Common Plots: Scatter plot, Hexbin plot, Jointplot.
- **Purpose**: Understand relationships (positive, negative, none).
- **Attributes**: `alpha`, `hue`, `style`, `size`.

```python
# Simple Example: Scatter Plot
x = np.random.rand(50)
y = 2*x + np.random.randn(50)*0.1
plt.scatter(x, y, color='red')
plt.title('Simple Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Detailed Example: Seaborn Jointplot
import pandas as pd

df = pd.DataFrame({
    'Height': np.random.normal(165, 10, 200),
    'Weight': np.random.normal(65, 15, 200)
})
sns.jointplot(x='Height', y='Weight', data=df, kind='scatter', color='purple')
plt.show()
```

## 3. Multivariate Plots

Multivariate plots analyze **three or more variables**.
- Common Plots: Pairplot, Heatmap.
- **Purpose**: Observe interactions between multiple variables.
- **Attributes**: `hue`, `palette`, `annot`, `cmap`.

```python
# Simple Example: Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, vars=['sepal_length','sepal_width','petal_length','petal_width'])
plt.show()
```

```python
# Detailed Example: Heatmap
# Select only the numeric columns before calculating correlation
numeric_iris = iris.select_dtypes(include=['number'])

# Calculate the correlation matrix on the numeric data
corr = numeric_iris.corr()

# Create the heatmap using the correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Iris Dataset)')
plt.show()
```

## 4. Correlation Coefficient

- Measures strength & direction of relationship between variables.
- Range: -1 (perfect negative) to +1 (perfect positive).
- Common method: Pearson correlation.

```python
import numpy as np
import pandas as pd

# Sample data
data = {
    'StudyHours': [2, 4, 6, 8, 10],
    'ExamScore': [50, 55, 65, 70, 90]
}
df = pd.DataFrame(data)
df_cor = df.corr()
# Calculate correlation matrix
print(df_cor)
```

```python
sns.heatmap(df_cor, annot=True, cmap='viridis')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Simple Example
x = np.random.rand(50)
y = x + np.random.randn(50)*0.1
corr = np.corrcoef(x, y)[0,1]
print('Correlation Coefficient:', corr)
```

```python
# Detailed Example with DataFrame
corr_matrix = df.corr()
print(corr_matrix)
sns.heatmap(corr_matrix, annot=True, cmap='viridis')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

## 5. Probability

- Probability = Likelihood of an event happening.
- Range: 0 (impossible) to 1 (certain).
- Examples: Coin toss, dice roll.
- In Python: `random` and `numpy` libraries.

```python
# Simple Example: Coin Toss
import random
outcomes = [random.choice(['Heads','Tails']) for _ in range(10)]
print('Coin Toss Outcomes:', outcomes)
```

```python
import random

# Simulate 1000 coin flips (Heads=1, Tails=0)
flips = [random.randint(0, 1) for _ in range(1000)]

# Calculate the probability of Heads (1)
prob_heads = sum(flips) / len(flips)

print(f"Total flips: {len(flips)}")
print(f"Number of Heads: {sum(flips)}")
print(f"Probability of Heads: {prob_heads}")
```

```python
# Sample Space (all possible outcomes)
sample_space = {1, 2, 3, 4, 5, 6}

# Event (outcomes where the number is even)
event_even = {2, 4, 6}

# Calculate Theoretical Probability: P(Event) = (Favorable Outcomes) / (Total Outcomes)
prob_even = len(event_even) / len(sample_space)

print(f"Probability of rolling an even number: {prob_even}")
```

```python
# Detailed Example: Dice Roll Simulation
dice_rolls = [random.randint(1,6) for _ in range(1000)]
sns.histplot(dice_rolls, bins=6, kde=False, color='orange')
plt.btitle('Dice Roll Distribution (1000 trials)')
plt.xlabel('Dice Face')
plt.ylabel('Frequency')
plt.show()
```
