# Statistics: Measures of Dispersion and Position with Visualizations

## 1. Variance

```python
import numpy as np
import matplotlib.pyplot as plt

# Example dataset
data = [2, 4, 6, 8]
mean = np.mean(data)
variance = np.var(data)  # population variance

print("Data:", data)
print("Mean:", mean)
print("Variance:", variance)

# Visualization
plt.figure(figsize=(6,4))
plt.bar(range(len(data)), data, color='skyblue')
plt.axhline(mean, color='red', linestyle='--', label=f'Mean = {mean}')
plt.title('Variance Visualization')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## 2. Standard Deviation

```python
std_dev = np.std(data)
print("Standard Deviation:", std_dev)

# Visualization
plt.figure(figsize=(6,4))
plt.hist(data, bins=4, color='lightgreen', edgecolor='black')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(mean+std_dev, color='blue', linestyle='--', label='+1 SD')
plt.axvline(mean-std_dev, color='blue', linestyle='--', label='-1 SD')
plt.title('Standard Deviation Visualization')
plt.legend()
plt.show()
```

## 3. Coefficient of Variation

```python
data2 = [10, 12, 15, 18, 20]
mean2 = np.mean(data2)
std2 = np.std(data2)
cv = (std2/mean2)*100
print("Data:", data2)
print("Mean =", mean2)
print("Standard Deviation =", std2)
print("Coefficient of Variation =", cv, '%')

# Visualization
plt.figure(figsize=(6,4))
plt.boxplot(data2)
plt.title('Coefficient of Variation Visualization (Spread)')
plt.show()
```

## 4. Skewness

```python
from scipy.stats import skew

salaries = [30, 35, 40, 45, 100]
print("Skewness =", skew(salaries))

# Visualization
plt.figure(figsize=(6,4))
plt.hist(salaries, bins=5, color='orange', edgecolor='black')
plt.title('Skewness Visualization')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, norm

# Set random seed for reproducibility
np.random.seed(42)

# 1. Generate data for different skewness types
# Negative Skew (left-skewed): take negative exponential
neg_skew = -np.random.exponential(scale=1, size=10000)

# Zero Skew (normal distribution)
zero_skew = np.random.normal(loc=0, scale=1, size=10000)

# Positive Skew (right-skewed): exponential distribution
pos_skew = np.random.exponential(scale=1, size=10000)

# 2. Calculate skewness values
print("Skewness Values:")
print("Negative Skew:", skew(neg_skew))
print("Zero Skew (Normal):", skew(zero_skew))
print("Positive Skew:", skew(pos_skew))

# 3. Plot histograms with PDFs
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Negative Skew
axes[0].hist(neg_skew, bins=50, density=True, alpha=0.6, color='skyblue')
xmin, xmax = axes[0].get_xlim()
x = np.linspace(xmin, xmax, 100)
axes[0].plot(x, norm.pdf(x, 0, 1), 'r--')
axes[0].set_title("Negative Skew (Left-skewed)")

# Zero Skew
axes[1].hist(zero_skew, bins=50, density=True, alpha=0.6, color='lightgreen')
xmin, xmax = axes[1].get_xlim()
x = np.linspace(xmin, xmax, 100)
axes[1].plot(x, norm.pdf(x, 0, 1), 'r--')
axes[1].set_title("Zero Skew (Normal)")

# Positive Skew
axes[2].hist(pos_skew, bins=50, density=True, alpha=0.6, color='salmon')
xmin, xmax = axes[2].get_xlim()
x = np.linspace(xmin, xmax, 100)
axes[2].plot(x, norm.pdf(x, 0, 1), 'r--')
axes[2].set_title("Positive Skew (Right-skewed)")

plt.tight_layout()
plt.show()
```

## 5. Kurtosis

```python
from scipy.stats import kurtosis

exam_scores = [50, 55, 60, 65, 70, 75, 80, 100]
print("Kurtosis =", kurtosis(exam_scores))

# Visualization
plt.figure(figsize=(6,4))
plt.hist(exam_scores, bins=6, color='purple', edgecolor='black')
plt.title('Kurtosis Visualization')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, norm

# Set random seed for reproducibility
np.random.seed(42)

# 1. Generate data for different kurtosis types
# Leptokurtic (heavy tails)  Laplace distribution
lepto = np.random.laplace(loc=0, scale=1, size=10000)

# Mesokurtic (normal distribution)
meso = np.random.normal(loc=0, scale=1, size=10000)

# Platykurtic (light tails)  Uniform distribution
platy = np.random.uniform(low=-2, high=2, size=10000)

# 2. Calculate kurtosis values
print("Kurtosis Values:")
print("Leptokurtic (Laplace):", kurtosis(lepto))
print("Mesokurtic (Normal):", kurtosis(meso))
print("Platykurtic (Uniform):", kurtosis(platy))

# 3. Plot histograms with PDFs
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Leptokurtic
axes[0].hist(lepto, bins=50, density=True, alpha=0.6, color='skyblue')
xmin, xmax = axes[0].get_xlim()
x = np.linspace(xmin, xmax, 100)
axes[0].plot(x, norm.pdf(x, 0, 1), 'r--')
axes[0].set_title("Leptokurtic (Laplace)")

# Mesokurtic
axes[1].hist(meso, bins=50, density=True, alpha=0.6, color='lightgreen')
xmin, xmax = axes[1].get_xlim()
x = np.linspace(xmin, xmax, 100)
axes[1].plot(x, norm.pdf(x, 0, 1), 'r--')
axes[1].set_title("Mesokurtic (Normal)")

# Platykurtic
axes[2].hist(platy, bins=50, density=True, alpha=0.6, color='salmon')
xmin, xmax = axes[2].get_xlim()
x = np.linspace(xmin, xmax, 100)
axes[2].plot(x, norm.pdf(x, 0, 1), 'r--')
axes[2].set_title("Platykurtic (Uniform)")

plt.tight_layout()
plt.show()
```

## 6. Z-Score

```python
from scipy.stats import zscore
import numpy as np
data = [2,4,6,8,10]
z_score = zscore(data)
print(z_score)
```

```python
x = 70
mean = 50
sd = 10

Z = (x - mean)/sd
print("Z-Score:", Z)

# Visualization
x_values = np.linspace(20, 80, 100)
y = (1/(sd*np.sqrt(2*np.pi))) * np.exp(-0.5*((x_values-mean)/sd)**2)

plt.figure(figsize=(6,4))
plt.plot(x_values, y, color='blue')
plt.axvline(x, color='red', linestyle='--', label=f'Score = {x}')
plt.title('Z-Score Visualization')
plt.legend()
plt.show()
```

## 7. Percentiles

```python
scores = [40, 50, 60, 70, 80, 90, 100]
p60 = np.percentile(scores, 60)
print("60th Percentile:", p60)

# Visualization
plt.figure(figsize=(6,4))
plt.boxplot(scores, vert=False)
plt.axvline(p60, color='red', linestyle='--', label='60th Percentile')
plt.title('Percentile Visualization')
plt.legend()
plt.show()
```

## 8. Quartiles

```python
data_q = [5, 7, 8, 12, 13, 14, 18, 21, 23]
Q1 = np.percentile(data_q, 25)
Q2 = np.percentile(data_q, 50)
Q3 = np.percentile(data_q, 75)
IQR = Q3 - Q1

print("Q1 =", Q1)
print("Q2 (Median) =", Q2)
print("Q3 =", Q3)
print("IQR =", IQR)

# Visualization
plt.figure(figsize=(6,4))
plt.boxplot(data_q)
plt.title('Quartile Visualization')
plt.show()
```
